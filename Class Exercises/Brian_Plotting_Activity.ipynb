### Line Plot

```python
x_data = [1,2,3,4,5,6,7,8,9]
y_data = [3,5,6,8,4,4,8,3,2]


plt.plot(x_data,y_data, 'r--', marker = '*' , label = 'Thing')
plt.title('Thing1')
plt.xlabel('X Thing')
plt.ylabel('Y Thing')
plt.legend()
plt.show
```
